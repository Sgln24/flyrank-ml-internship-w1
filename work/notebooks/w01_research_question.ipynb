## 2. The Question: Decision, Action, Cost of a Wrong Call

### Research Question

Which content pages should be reviewed first for refresh, expansion, protection, pruning, or monitoring based on observable search and engagement signals?

### Decision

The decision is determining which pages deserve the highest review priority.

### Unit of Analysis

One row represents one content page.

### Output

A ranked priority score indicating which pages should be reviewed first.

### Who Uses the Output?

SEO specialists and content editors.

### Action

The review team examines the highest-ranked pages first and decides whether each page should be refreshed, expanded, monitored, or otherwise improved.

### Cost of a Wrong Recommendation

If the ranking is incorrect, valuable editorial time may be spent reviewing low-impact pages while higher-impact opportunities are delayed. This reduces the efficiency of the content optimization process.

### Why Data / Machine Learning?

Search performance depends on many interacting signals including visibility, freshness, traffic, engagement, and content characteristics. These relationships are difficult to capture using a single rule-based approach. Machine learning can combine multiple observable signals to prioritize review candidates more effectively while remaining a decision-support tool rather than replacing human judgment.

## 3. Quick Look at the Data

The starter dataset contains **30,000 content items** described by **44 features**, providing a sufficiently large sample for exploratory analysis and ranking experiments.

The data covers **32 anonymized clients**, which introduces variation across different websites and helps evaluate whether prioritization methods generalize beyond a single client.

Most of the dataset consists of **keyword articles (27,207 pages)**, followed by **Feedly articles (2,096 pages)** and **comparison articles (697 pages)**. This suggests that keyword-focused content dominates the available inventory.

The average number of impressions over the previous 90 days is **5,200.37**, indicating that many pages have meaningful search visibility.

The average click-through rate (CTR) is approx. **0.51%** (note that CTR is stored as a percentage value, so 0.51 represents **0.51%**, not 51%).

After excluding rows where no ranking data is available (`avg_position = 0`), the average Google search position is **17.03**, meaning that many pages are already visible in search results but still have room for improvement.

These observations support the choice of the **Refresh / Content Opportunity Scoring** lane because the dataset contains enough variation in visibility, ranking position, and engagement to prioritize which pages should be reviewed first.
