## 1. My lane as an ML task (type)

## Task Type

This project is a scoring (regression-style) machine learning problem.

The goal is to assign each content page a priority score that estimates how valuable it would be to review that page for content refresh.

Rather than making a yes/no decision, the model produces a score that allows pages to be ranked from highest to lowest review priority.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd

url = "https://raw.githubusercontent.com/Sgln24/flyrank-ml-internship-w1/main/data/raw/content_refresh_anonymized.csv"

df = pd.read_csv(url)

print("Rows:", len(df))
print("Columns:", len(df.columns))
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

### Why ML Beats a Fixed Rule

A fixed rule, such as prioritizing only pages with the lowest CTR or the highest impressions, ignores many other important signals.

Machine learning can combine multiple observable features, such as impressions, clicks, CTR, average position, and content freshness, to identify complex patterns that are difficult to capture with simple if-statements.

This provides a more balanced and evidence-based priority score while still supporting human decision making.
